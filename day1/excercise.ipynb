```python
# use use bargraph when we have to compare the categories

import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'titanic' is your DataFrame and 'Sex' is the column
sex_counts = pd.value_counts(titanic['Sex'])

print(sex_counts)

plt.bar(sex_counts.index, sex_counts.values)
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Sex')
plt.show()
```

```python
# divides that data into 25% of the distribution, or we compare the quartiles of two categories
survived = titanic[titanic["Survived"] == 1]["Age"]
notSurvived = titanic[titanic["Survived"] == 0]["Age"]

plt.boxplot([survived, notSurvived], ["survived", "not survived"])
plt.ylabel("age")
plt.title("Survived vs Not Survived")
plt.show()

```
